```python
data = open('/home/shobhit/NewDataset/UCI HAR Dataset/train/X_train.txt')
```

```python
lines = data.readlines()
len(lines)
```

```python
lines[1]
```

```python
new_lines=[]
for i,line in enumerate(lines):
    temp = line.strip('\n')
    temp = temp.strip('  ')
    temp = re.sub("\s+",",",temp.strip())
    a=[float(e) for e in temp.split(',')]
    new_lines.append(a)
```

```python
z=np.array(new_lines)
z.shape
```

**Dimentionality Reduction with PCA**

```python
pca = PCA(n_components=4)
pca.fit(z)
```

**Top 4 Dimentions and their respection variance ratio**

```python
print(pca.explained_variance_ratio_)
```

```python
sample = pca.transform(z)
sample[9]
```

**Scatter Plot**

```python
area = np.pi*3
plt.scatter(sample[:,0],sample[:,1],s=area,c='red',alpha=0.5)
plt.show()
```
